## Doubly Linked List (DLL)

### Insertion at the beginning, at end and at a specific position in Doubly Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
class DLL:
    def __init__(self):
        self.head = None

    # Insert at the beginning
    def insert_at_begin(self, data):
        nb = Node(data)
        if self.head is None:
            self.head = nb
        else:
            nb.next = self.head
            self.head.prev = nb
            self.head = nb

    # Insert at the end
    def insert_at_end(self, data):
        ne = Node(data)
        if self.head is None:
            self.head = ne
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.next = ne
            ne.prev = temp         
    def insert_at_position(self, position, data):
        np = Node(data)
        if position == 1:
            L.insert_at_begin(data)
        else:
            temp = self.head
            for i in range(1, position-1):
                temp = temp.next
            if temp is None:
                print("position out of bound")
            else:
                np.next = temp.next
                np.prev = temp
                if temp.next is not None:
                    temp.next.prev = np
                temp.next = np    
    def display(self):
        if self.head is None:
            print("linked list is empty")
        else:
            a = self.head
            while a is not None:
                print(a.data, end=" --> ")
                a = a.next
            print("None")         
    def backward_traversal(self):
        if self.head is None:
            print("linked list is empty")
        else:
            a = self.head
            while a.next is not None:
                a = a.next
            while a is not None:
                print(a.data, end=" --> ")
                a = a.prev
            print("None")

L = DLL()
L.insert_at_begin(10)
L.insert_at_begin(20)      
print("After inserting data at begin")       
L.display()
print("\nAfter inserting data at end")   
L.insert_at_end(99)           
L.display()
print("\nBackward Traversal :")   
L.backward_traversal()
print("\nAfter inserting data at position=4")   
L.insert_at_position(4, 169)           
L.display()
print("\nBackward Traversal :")   
L.backward_traversal()
```

## Circular Linked List

### Insertion, deletion, display in Circular Linked list using (single pointer)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class CLL:
    def __init__(self):
        self.head = None

    def insert_at_begin(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
        else:
            new_node.next = self.head
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            self.head = new_node
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_node
            new_node.next = self.head 
            
    def insert_at_position(self, position, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
        else:
            if position ==1:
                L.insert_at_begin(data)
            else:
                temp = self.head
                for i in range(1, position-1):
                    temp = temp.next
            new_node.next = temp.next 
            temp.next = new_node
            if new_node.next == self.head:
                new_node.next = self.head
            
    def delete_at_begin(self):
        if self.head is None:
            print("Circular linked list is empty")
        elif self.head.next == self.head:  
            self.head = None
        else:
            a = self.head
            while a.next != self.head:
                a = a.next
            self.head = self.head.next
            a.next = self.head
            self.tail = a 
            a = None
    def delete_at_end(self):
        if self.head is None:
            print("Circular linked list is empty")
        elif self.head.next == self.head:  
            self.head = None
        else:
            a = self.head
            while a.next.next != self.head:
                a = a.next
            a.next = self.head

    def delete_at_position(self, position):
        if self.head is None:
            print("cll is empty")
        else:
            if position ==1:
                L.delete_at_begin(self)
            else:
                temp = self.head
                for i in range(1, position-1):
                    temp = temp.next
            if temp.next == self.head:
                print("position out of bounds")
            else:
                node_to_delete = temp.next
                temp.next = node_to_delete.next
                node_to_delete = None
                   
    def display(self):
        if self.head is None:
            print("Circular linked list is empty")
        else:
            a = self.head
            print(a.data, end=" --> ")
            while a.next != self.head:
                a = a.next
                print(a.data, end=" --> ")
            print(a.next.data) 

L = CLL()
L.insert_at_begin(10)
L.insert_at_begin(20)
L.insert_at_begin(30)
L.insert_at_begin(40)
print("After inserting data at begin")
L.display()
print("After deleting data at end")
L.delete_at_end()
L.display() 
print("After inserting data at end")
L.insert_at_end(111)
L.display()
print("After inserting data at position=3")
L.insert_at_position(3, 188)
L.display()
print("After deleting data at begin")
L.delete_at_begin()
L.display() 
print("After inserting data at end")
L.insert_at_end(1000)
L.display()
print("After deleting data at position=4")
L.delete_at_position(4)
L.display()
```

### Node deletion in Circular Linked List using (head, tail) pointer

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CLL:
    def __init__(self):
        self.head = None
        self.tail = None

    def insert_at_begin(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            self.tail.next = self.head
        else:
            new_node.next = self.head
            self.tail.next = new_node
            self.head = new_node

    def delete_at_begin(self):
        if self.head is None:
            print("Circular linked list is empty")
        elif self.head == self.tail:  
            self.head = None
            self.tail = None
        else:
            temp = self.head
            self.head = temp.next
            self.tail.next = self.head      # Update tail to point to the new head
            temp = None                     # Free the old head
    def delete_at_end(self):
        if self.head is None:
            print("Circular linked list is empty")
        elif self.head == self.tail:  
            self.head = None
            self.tail = None
        else:
            a = self.head
            while a.next != self.tail:
                a = a.next
            self.tail = None
            self.tail = a 
            self.tail.next = self.head

    def display(self):
        if self.head is None:
            print("Circular linked list is empty")
        else:
            a = self.head
            print(a.data, end=" --> ")
            while a.next != self.head:
                a = a.next
                print(a.data, end=" --> ")
            print(a.next.data) 

L = CLL()
L.insert_at_begin(10)
L.insert_at_begin(20)
L.insert_at_begin(30)
L.insert_at_begin(40)
print("After inserting data at end")
L.display()

print("\nAfter deleting data at begin")
L.delete_at_begin()
L.display()     

print("\nAfter deleting data at end")
L.delete_at_end()
L.display() 
```

## Header Linked List 

### Insertion, deletion and display in Header Linked List

```python
class Node:
    def __init__(self, data=None):        # data=None for header node
        self.data = data
        self.next = None
class HeaderLinkedList:
    def __init__(self):
        self.header = Node()              # Header node with no data    
    def insert_at_begin(self, data):
        new_node = Node(data)
        new_node.next = self.header.next
        self.header.next = new_node
    def insert_at_end(self, data):
        new_node = Node(data)
        temp = self.header
        while temp.next is not None:
            temp = temp.next
        temp.next = new_node      
    def insert_at_position(self, position, data):
        new_node = Node(data)
        if position == 1:
            new_node.next = self.header.next
            self.header.next = new_node
        else:
            temp = self.header.next
            for i in range(1, position-1):
                temp = temp.next
            new_node.next = temp.next
            temp.next = new_node

    def delete_at_begin(self):
        temp = self.header
        if temp.next is None:
            print("The list is empty")
        else:
            first_node = self.header.next
            self.header.next = first_node.next
            del first_node
    def delete_at_end(self):
        temp = self.header
        if temp.next is None:
            print("The list is empty")
            return
        while temp.next.next is not None:
            temp = temp.next
        last_node = temp.next
        temp.next = None
        del last_node      
    def delete_at_position(self, position):
        temp1 = self.header
        if temp1.next is None:
            print("The list is empty")
            return
        else:
            temp2 = self.header.next
            for i in range(1, position-1):
                temp2 = temp2.next
            node_to_delete = temp2.next
            temp2.next = node_to_delete.next
            del node_to_delete

    def display(self):
        if self.header.next is None:
            print("Linked list is empty")
        else:
            a = self.header.next
            while a is not None:
                print(a.data, end=" --> ")
                a = a.next
            print("None")

L = HeaderLinkedList()
L.insert_at_begin(10)
L.insert_at_begin(20)
L.insert_at_begin(30)
L.insert_at_begin(40)
print("After inserting data at beginning:")
L.display()
print("\nAfter deleting data at end:")
L.delete_at_end()
L.display()
print("\nAfter inserting data at end:")
L.insert_at_end(50)
L.display()
print("\nAfter deleting data at begin:")
L.delete_at_begin()
L.display()
print("\nAfter inserting data at position=3")
L.insert_at_position(3, 100)
L.display()
print("\nAfter deleting data at position=2")
L.delete_at_position(2)
L.display()
```
